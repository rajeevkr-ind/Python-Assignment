```python
# Write a code to count the number of vowels in a string.

# Vowels = "aeiou"
def count_vowels(input_string):
    # Define a set of vowels
    vowels = "aeiouAEIOU"
    
    # Initialize a counter for vowels
    count = 0
    
    # Iterate through each character in the string
    for char in input_string:
        if char in vowels:
            count += 1
            
    return count

# Example usage
input_string = "Hello, World!"
vowel_count = count_vowels(input_string)
print(f"The number of vowels in '{input_string}' is: {vowel_count}")
The number of vowels in 'Hello, World!' is: 3
```

```python
# Write a code to shuffle a given list randomly without using any built-in shuffle functions.

# import random
random
<module 'random' from '/opt/conda/lib/python3.10/random.py'>

def shuffle_list(input_list):
    # Create a copy of the input list to avoid modifying the original
    shuffled_list = input_list[:]
    n = len(shuffled_list)
    
    # Implement the Fisher-Yates shuffle
    for i in range(n - 1, 0, -1):
        # Generate a random index from 0 to i
        j = random.randint(0, i)
        
        # Swap elements at indices i and j
        shuffled_list[i], shuffled_list[j] = shuffled_list[j], shuffled_list[i]

    return shuffled_list

# Example usage
original_list = [1, 2, 3, 4, 5]
shuffled = shuffle_list(original_list)
print("Original list:", original_list)
print("Shuffled list:", shuffled)
```
